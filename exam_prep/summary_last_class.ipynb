## Reinforcement Learning

Explain the bellman equation conceptionally and mathematically. <br>
- The bellman equation is used to update state values and action values over different iterating loops

What is the difference between the action value and the state value? 

**State Value V(s):**
- state value represents the value of being in a particular state
- estimates the expected return strating from tha stat and following the current policy thereafter
- state value describes the overall goodness or desirability of a state, irrespective of the specific actions taken in that state


**Action Value Q(s, a):**
- action value represents the value of taking a particular action in a given state
- it estimates the expected return from starting from a given state, taking a specific action and following the current policy thereafter
- action value considers both the immediate reward obtained by taking the action and the expected return from the subsequent states


Explain N-armed bandit conceptionally and mathematically. <br>
- N-armed bandit help to decide whether to better explore or exploit. They can exploit with the current estimate of action values following the greedy policy. Or they can explore improving the estimates of the action values through non-greedy action.
- the tradeoff depends on the number of remaining steps and the values and uncertainty of the estimates


What is the Policy Improvement Theorem? <br>
- Always following π' is better than following π if it is better to select next action from π' and thereafter π than to follow π all the time.

What is the policy improvement?
- greedy policy π' that selects the action that looks best after one step lookahead according to policy π

Explain what happens in the value iteration?

- iterate through states using the Bellman optimal equation for state value (select best action following the best value )
- calculate the difference between previous value state and current value state
- stop if difference is a small positive number
- output deterministic policy such that
$$\pi(s) = \argmax_a  \sum_{s',r}p(s',r |s,a)[r + \gamma V(s')]$$

What is the Monte Carlo Methods used for in reinforcement learning?
- estimates value function from experience i.e. averging sample returns

What is the difference between off-policy and on-policy methods?

**on-policy**
- evaluate / improve the policy that is used to make decisions
- policy being learned and policy being followed are the same
- data collected from the environment is used to both make decisions and update the policy.
- soft but gradually to deterministic optimal policy

**off-policy**
- Off-policy methods update the policy based on data collected by a different policy (different from the one used to generating experiencing the environment(sampling) ).
- the agent learns from the experience of other policies.
- data collected from the behaviour policy is used to update the target policy
- the behaviour policy can explore more widely that the target policy

## Diffusion

What is the ELBO and why is it important?
- ELBO is the evidence lower bound
- we want to maximise the log-likelihood
- instead of maximising log-likelihood we can use ELBO as proxy
- lower bound equivalent to evidence if parametrized optimally

What is the difference between the Markovian-hierarchical, Autoencoder and diffusion model?

**VAE**
- learn compressed latent representations
- consists of encoder that maps to the latent space and decoder that reconstructs from the latent space
- VAE employ variational inference to optimise model parameters and learn latent distribution
- VAEs generate new samples by sampling from learned latent space

**MH-VAE**
- Combination of VAE and hierarchical models
- Aim to model high dimensional data with complex dependencies incorporating the markov property
- latent space structured hierarchical different levels of abstraction

**Diffusion Models**
- data generation process as a continous-time diffusion process
- transformation of an initial noise distribution over time to generate samples
- uses denoising score matching to estimate the diffusion process 

Explain VAE:

- we want to maximise the reconstruction term (how likely we can reconstruct our original x from the latent space z)
- we want to minimise prior matching term (model learns to balance between accurate data reconstruction and adherence to the prior distribution i.e. multivariate gaussian)
